# **Dataset**

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer

# Upload the breast cancer dataset from scikit-learn
breastcancer = load_breast_cancer()

# Input for machine learning
X = breastcancer.data
# Target output from machine learning
y = breastcancer.target

# Converts feature and target data into a DataFrame
df_X = pd.DataFrame(X, columns=breastcancer.feature_names)
df_y = pd.Series(y, name='target')
```

```python
# Displaying features from the dataset
df_X
```

```python
# Displaying target from the dataset
df_y
```

```python
# Merging features and target into a single DataFrame
df = pd.concat([df_X, df_y], axis=1)

#Viewing the top 10 rows of the dataset
df.head(10)
```

```python
#Displaying a summary of information about the DataFrame
df.info()
```

```python
#Viewing unique values in the 'target' column from the DataFrame
df['target'].unique()
```

```python
# Displaying a summary of descriptive statistics for numerical columns in the DataFrame
df.describe()
```

# **Split Data**

```python
from sklearn.model_selection import train_test_split

# Splitting the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(df_X, df_y, test_size=0.2, random_state=42)
```

# **Train the Model**

```python
from sklearn.ensemble import RandomForestClassifier

# Creating and training a Random Forest model
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)
```

# **Predict & Evaluate**

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Predict and evaluate
y_pred = model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)

print("Laporan Klasifikasi:")
print(f"Akurasi: {accuracy * 100:.2f}%")
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred)

# Plot Confusion Matrix using seaborn
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues",
            xticklabels=breastcancer.target_names, yticklabels=breastcancer.target_names)
plt.xlabel("Prediksi")
plt.ylabel("Aktual")
plt.title("Confusion Matrix")
plt.show()
```

```python
from sklearn.metrics import accuracy_score, classification_report

# Displaying classification report
print(classification_report(y_test, y_pred, target_names=breastcancer.target_names))
```
